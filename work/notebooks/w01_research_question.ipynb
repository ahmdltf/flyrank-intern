## 3. Quick look at the data (2-3 real numbers)

*2–3 Real Numbers from Data:*

*Striking Distance Opportunity: There are 9,069 pages in the striking distance position (rank 4–20) with high impressions (>1,000 impressions/90 days). This is the main target for CTR optimization.*

*Underperforming CTR Gap: The current average CTR is 0.51%, but 50% of the data (median CTR) only reaches 0.07%. This proves the majority of pages are performing well below their potential.*

*Volume vs Conversion Opportunity: The average page received 5,200 impressions (90 days) but only generated 16.1 clicks, indicating a large gap that can be improved through engagement and CTR improvements.*
